```python
import csv

# Define a class Person which takes house-id, age and gender as constructor arguments.
# Define a method income() and return-zero. By default, people do not earn.
# Define a method expenditure() and return value as explained above.
class Person:
    def __init__(self, household, age, gender):
        self.household = household;
        self.age = age;
        self.__gender = gender;

    def income(self):
        return 0

    def expenditure(self):
        if self.__gender == 'm':
            return 500 + 5 * self.age
        if self.__gender == 'f':
            return 400 + 7 * self.age
        raise Exception('Unknown gender ' + self.__gender) 

# Next define a class Household which taked household id as the argument and let
# it have a property members_list which is a list of Person instances.
# Implement a method num_adults() and return the number of adults > 18 years.
# Implement a method savings() which totals over member's income and expenditure.
class HouseHold:
    def __init__(self, id):
        self.__id = id
        self.__members_list = []

    def add(self, person):
        self.__members_list.append(person)

    def size(self):
        return len(self.__members_list)
    
    def adults(self):
        return len([1 for m in self.__members_list if m.age >= 18])

    def savings(self):
        savings = 0
        for member in self.__members_list:
            savings += member.income() - member.expenditure()
        return savings

# Define following sub-classes of Person - BusinessPerson, Employed, Politician
# and Student.
# Implement income() of BusinessPerson, Employed, Politician as explained above.
# For Student - you have two options to model education expenditure
#     a. Let income() return a negative amount.
#     b. Override expenditure() and return super.expenditure() + education-expense.
#
# Now since income of Business people depends on the number of adumts in the
# household, the income method should take an argument - an instance of household.
# Go back and change the method signature to make income() take an argument, and
# change HouseHold.savings() method to pass 'self' as an argument to Person.income()
#
# (Note: It is also possible to do this without changing method signature, for that
# you will have to change the constructor argument of BusinessPerson to take the
# Household instance as a member attribute. This means one will have to instanciate
# Household before instanciating BusinessPerson).

class BusinessPerson(Person):
    def income(self):
        return 2000 * self.household.adults()

class Employee(Person):
    def income(self):
        if (age <= 60):
            return 1500 * self.age
        return 5000

class Student(Person):
    def income(self):
        return - (1000 + 10* self.age)

class Politician(Person):
    def income(self):
        return 2500 * (self.age - 20)

# Let us create a map (a dictionary called a map datastructure) of
# household-id -> Household.

household_by_id = {}

# Read the csv data file
# For every line read, check if HouseHold already exists in the dictionary.
# If not, create a HouseHold instance and put it in the dictionary, else use the
# existing one.
with open('/home/data/household.csv','r') as f:
    data = csv.reader(f)
    next(data)  # Skip headers
    for d in data:
#         print(d)
        h_id = d[0]
        age = int(d[1])
        gender = d[2]
        if h_id in household_by_id:
            household = household_by_id[h_id]
        else:
            household = HouseHold(h_id)
            household_by_id[h_id] = household

        # Create a correct person instance depending on the profession.
        if d[3] == 'Business':
            person = BusinessPerson(household, age, gender)
        elif d[3] == 'Employed':
            person = Employee(household, age, gender)
        elif d[3] == 'Politician':
            person = Person(household, age, gender)
        elif age >= 4 and age <= 21:
            person = Student(household, age, gender)
        else:
            person = Person(household, age, gender)

        household.add(person)



household_savings = []
# Iterate through each household, computing its savings and adding it to the array.

for household in household_by_id.values():
    household_savings.append(household.savings())

# Sort the array and find the median.
sorted_savings = sorted(household_savings)
print("Median", sorted_savings[len(sorted_savings) // 2])
```
